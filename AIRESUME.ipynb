```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download necessary NLP data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))

# Function to clean text
def preprocess_text(text):
    if pd.isnull(text):
        return ""  # Handle missing values
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\W', ' ', text)  # Remove special characters
    words = word_tokenize(text)  # Tokenize words
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]  # Remove stopwords & lemmatize
    return " ".join(words)

# Apply preprocessing to resume and job descriptions
df["cleaned_resume"] = df["career_objective"].apply(preprocess_text)  # Assuming career_objective contains resume details
df["cleaned_job_description"] = df["skills_required"].apply(preprocess_text)  # Assuming skills_required contains job details

# Check cleaned data
df[["cleaned_resume", "cleaned_job_description"]].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform resume and job descriptions
resume_tfidf = tfidf_vectorizer.fit_transform(df["cleaned_resume"])
job_tfidf = tfidf_vectorizer.transform(df["cleaned_job_description"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute cosine similarity
match_scores = cosine_similarity(resume_tfidf, job_tfidf)

# Store match score in the dataset
df["matched_score"] = match_scores.diagonal() * 100  # Convert to percentage
```

```python
# Sort candidates by match score
df_sorted = df.sort_values(by="matched_score", ascending=False)

# Display top 10 candidates
df_sorted[["career_objective", "skills_required", "matched_score"]].head(10)
```
